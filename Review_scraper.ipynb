Extraire les reviews 

```python
review_list = []
```

```python
for i in range(1, len(reviews)):
    review = driver.find_element(By.XPATH,f'//div[@id="reviews-container"]')
    # Extract the rating and review text
    Name = review.find_element(By.XPATH, f"//div[@id='review-{i}']//div[@class='EhHd7']//span[@class='dls-1n7v84y']").text
    Date = review.find_element(By.XPATH, f"//div[@id='review-{i}']//div[@class='VqCvj']//span[@class='OOO01 dls-1n7v84y']").text
    Rate = review.find_element(By.XPATH, f"//div[@id='review-{i}']//span[@class='F0_PJ']")
    style_value = Rate.get_attribute("style")
    Rate = str(float(style_value.split(";")[0].split(":")[1].strip().replace("%", ""))*0.05) + '/5'
    Subject = review.find_element(By.XPATH, f"//div[@id='review-{i}']//span[@class='dls-1n7v84y']//strong[@class='gBRHe']").text
    Text = review.find_element(By.XPATH, f"//div[@id='review-{i}']//div[1]/div/div[2]").text
    
    try:
       People_liked = review.find_element(By.XPATH, f"//div[@id='review-{i}']//div[@class='pPLTf']//span[@class='yLx9a dls-1n7v84y']").text
    except:
       People_liked = "-"
       
    try:
       Type_purchase = review.find_element(By.XPATH, f"//div[@id='review-{i}']//div[@class='Xjtf_']//p[@class='FVG_7 dls-1n7v84y'][1]").text
       
    except:
       Type_purchase = "Not verified"
       
    try:
       Picture_review = review.find_element(By.XPATH, f'//div[@id="review-{i}"]//div[@class="LutYS Xdo9E idJS8 W04Ff"]//img')
       image_url = Picture_review.get_attribute('src')
    except:
       image_url = "-"
    

    
    # Add the rating and review text to a dictionary
    review_dict = {
        'Name': Name,
        'Date': Date,
        'Rate': Rate,
        'Subject': Subject,
        'Text': Text,
        'People_liked': People_liked,
        'Type_purchase' : Type_purchase,
        'Image_review' : image_url
    }
    
    # Add the dictionary to a list
    review_list.append(review_dict)
```

Extraire les images des reviwes

```python
number_scrolls = 2


for i in range(1, number_scrolls):
    wait=WebDriverWait(driver,1)
    
    images = driver.find_elements(By.XPATH, '//li[@class="sjC0e"]//img')
    image_srcs = [image.get_attribute('src') for image in images]
    df = pd.DataFrame({'image_url': image_srcs})

    more_scrolls_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '//button[@class="pUkmH piydo"]')))
    more_scrolls_button.click()

    # Wait for the reviews to load
    time.sleep(5)
```
